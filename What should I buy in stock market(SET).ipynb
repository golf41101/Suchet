เนื่องจากข้อมูลหุ้นใน Yahoo จะมีคำว่า .BK ต่อท้าย ดังนั้นเราจึงต้องเพิ่ม +'.BK' เข้าไปในหุ้นทุกตัว

```python
from pandas_datareader import data as wb
```

```python
tickers = [stock]
mydata = pd.DataFrame()
for t in tickers :
    mydata = wb.DataReader(t, data_source='yahoo', start = '2019-07-01', end = '2020-06-30')['Adj Close']
```

```python
# จากการดึงข้อมูลไม่พบ หุ้น SAWAD SCB SCC TOA เราจะทำการดึงหุ้นเหล่านี้เพิ่มในภายหลังครับ
mydata.head()
```

```python
# ลบช้อมูล NA ที่ดึงไม่ได้ และไม่มีข้อมูลทิ้ง หุ้น AWC ขายใน SET วันแรก 10/10/62 และ หุ้น CRC ขายใน SET วันแรก 20/02/63 
# จึงทำให้มีข้อมูลไม่เพียงพอต่อการคำนวณ เราจึงจะตัด 2 ตัวนี้ทิ้งไป
mydata = mydata.dropna(axis='columns')
```

```python
mydata.tail()
```

```python
# ดึงข้อมูลเพิ่มใน หุ้น SAWAD SCB SCC TOA เราจะทำการดึงหุ้นเหล่านี้เพิ่มในภายหลังครับ
tickers_2 = [['SAWAD.BK','SCB.BK','SCC.BK','TOA.BK']]
mydata2 = pd.DataFrame()
for t in tickers_2 :
    mydata2 = wb.DataReader(t, data_source='yahoo', start = '2019-07-01', end = '2020-06-30')['Adj Close']
```

```python
mydata2.head()
```

```python
# ทำการรวมข้อมูล จะได้ทั้งหมด 48 ตัว ตัด 2 ตัวทิ้งเพราะข้อมูลไม่เพียงพอต่อสมมติฐานตั้งต้นครับ
alldata_1 = pd.merge(mydata,mydata2,on='Date')
```

```python
alldata_1.head()
```

```python
# หรืออีกวิธีในการรวมข้อมูลตาราง
alldata_2= pd.concat([mydata, mydata2], axis=1)
```

```python
alldata_2.head()
```

```python
from statistics import stdev
```

```python
SD_SET50 = []
for i in alldata_1 :
    SD_SET50.append(stdev(alldata_1[i]))
```

```python
# เช็คว่ารันครบมั้ยครับ
len(SD_SET50)
```

```python
# ทำการ Transpose ข้อมูล (สลับแถวสลับหลัก) เพื่อให้ดูง่ายขึ้นครับ
newdata = alldata_1.T
```

```python
newdata.head()
```

```python
newdata['SD'] = SD_SET50
```

```python
newdata['SD'].head()
```

```python
(alldata_1['ADVANC.BK']['2020-06-30'] - alldata_1['ADVANC.BK']['2019-07-01'])/alldata_1['ADVANC.BK']['2019-07-01']
```

```python
growth_price = [] # growth_price as gp
for gp in alldata_1 :
    growth_price.append(((alldata_1[gp]['2020-06-30']-alldata_1[gp]['2019-07-01'])/alldata_1[gp]['2019-07-01']*100))
    
```

```python
growth_price 
```

```python
newdata['Growth Price(%)'] = growth_price
```

```python
newdata[['SD','Growth Price(%)']]
```

```python
# มาถึงจุดที่เราต้องการละนะครับ หาความผันผวนต่อการเปลี่ยนแปลงของราคา ซึงเราต้องการค่าตัวนี้มากสุด 
# เพื่อ ที่จะเลือกซื้อหุ้นที่ผันผวนน้อยและมีการเติบโตที่เป็นบวกมากที่สุด
newdata['GrowthperSD'] = (newdata['Growth Price(%)']/newdata['SD'])
```

```python
# มานับกันหน่อยว่าครบมั้ย
len(newdata['GrowthperSD'])
```

```python
newdata['GrowthperSD'].head()
```

```python
focus_data = newdata[['Growth Price(%)','SD','GrowthperSD']]
```

```python
focus_data.head()
```

```python
# ผมต้องการที่จะแยกหัวตารางคำว่า Date(พอดีตอนใช้ Datareader ดึงข้อมูลมาเลยติดคำว่า Date มาด้วยครับ) กับ Symbols ออกจากกัน
focus_data = focus_data.reset_index()
```

```python
# แต่ตอนนี้ Date เป็นหัวตารางใน Index แทน
focus_data.head()
```

```python
# คร่าวนี้ผมจะเปลี่ยน ชื่อหัวตารางของ Index ซึง Date ให้ไม่มีชื่อ จึงใส่ค่า None ลงไปครับ หรือจะพิม '' ก็ได้เช่นกันครับ
focus_data = focus_data.rename_axis(None, axis='columns')
```
